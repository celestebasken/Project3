```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded dataset (fix path as needed)
df = pd.read_csv("WLD_RTFP_country_2023-10-02.csv")

# Convert 'date' to datetime format and extract the year
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

# Filter for Niger and Nigeria in 2020
df_filtered = df[(df['country'].isin(['Niger', 'Nigeria'])) & (df['year'] == 2020)]

df_filtered
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Filter for Niger and Nigeria in 2020
df_filtered = df[(df['country'].isin(['Niger', 'Nigeria'])) & (df['year'] == 2020)]

# Plot inflation trend for each country by month
plt.figure(figsize=(10, 6))
for country in df_filtered['country'].unique():
    monthly_avg = df_filtered[df_filtered['country'] == country].groupby('month')['Inflation'].mean()
    plt.plot(monthly_avg.index, monthly_avg.values, marker='o', label=country)

plt.title("Monthly Average Inflation in Niger and Nigeria (2020)")
plt.xlabel("Month")
plt.ylabel("Average Inflation")
plt.xticks(range(1, 13))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Ethiopia Inflation data from the UN
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
ethiopia_inflation = [18.7, 21.8, 22.6, 22.9, 19.8, 21.6, 22.3, 20, 18.7, 19.3, 19, 18.2]

plt.figure(figsize=(10, 6))
plt.plot(months, ethiopia_inflation, marker='o', linestyle='-', color='blue')

plt.title('Ethiopia Monthly Inflation')
plt.xlabel('Month')
plt.ylabel('Inflation Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
